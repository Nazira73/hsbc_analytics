```python
# plot year sale

plt.figure(figsize = (8,3))
plt.style.use('seaborn-v0_8-whitegrid')
data = df.groupby(by = df['Order Date'].dt.to_period('Y'))['Sales'].sum()
time = data.index.to_timestamp()
# time = [str(year) for year in data.index]
sales = data.values

plt.plot(time, sales,color = 'pink', marker = "o", linestyle = "--", linewidth = 2, markersize = 5,markerfacecolor = 'blue')
plt.title("Sales", fontsize = 25)
plt.xlabel("Time", fontsize = 10)
plt.ylabel("Sales", fontsize = 10)
plt.legend(["Sales Trend"], loc = 'lower right')
plt.show()
```

```python
# plot monthly sale

data = df.groupby(by = df['Order Date'].dt.to_period('M'))['Sales'].sum()
time = data.index.to_timestamp()
sales = data.values
custome_x_ticks = [time[i] for i in range(0,len(time),3)]

plt.figure(figsize = (16,5))
plt.plot(time, sales,color = 'pink', marker = "o", linestyle = "--", linewidth = 2, markersize = 5,markerfacecolor = 'blue')
plt.xticks(custome_x_ticks, rotation = 80)
plt.show()
```

```python
# plot quaterly sale

data = df.groupby(by = df['Order Date'].dt.to_period('Q'))['Sales'].sum()
time = data.index.astype('str')
sales = data.values

plt.figure(figsize = (16,5))
plt.plot(time, sales,color = 'pink', marker = "o", linestyle = "--", linewidth = 2, markersize = 5,markerfacecolor = 'blue')
plt.xticks(time, rotation = 80)
plt.show()
```

```python
# Change Dateformat pf x-axis

import matplotlib.dates as mdates

data = df.groupby(by = df['Order Date'].dt.to_period('M'))['Sales'].sum()
time = [date.strftime('%Y-%m-%d') for date in data.index.to_timestamp()]
sales = data.values

plt.figure(figsize = (15,5))
plt.plot(time, sales,color = 'pink', marker = "o", linestyle = "--", linewidth = 2, markersize = 5,markerfacecolor = 'blue')
plt.xticks(time, rotation = 90)                                      
plt.show()
```

```python
# Plot sales category wise

data = df.groupby(by = [df['Order Date'].dt.to_period('M'),'Category'])['Sales'].sum().reset_index().set_index('Order Date')
time = data.index.to_timestamp()

plt.figure(figsize = (15,5))

colors = ['blue','yellow','green']
linestyles = ['--','dotted','-.']
categories = data['Category'].unique()

for cat,color,ls in zip(categories, colors,linestyles):
    cat_data = data[data['Category'] == cat].sort_index()
    time = cat_data.index.to_timestamp()
    sales = cat_data['Sales']
    plt.plot(time, sales,color = color, label = cat, linestyle = ls, linewidth = 3)
    
plt.legend(loc = 'upper left')
plt.show()
```

```python
bins = 5

cmap = plt.get_cmap('Pastel1')
colors = cmap(np.linspace(0, 1, bins))

counts,bins,patches = plt.hist(df['Quantity'], rwidth = 0.9, bins = bins)
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.xticks(bins)

for i in range(len(patches)):
    patches[i].set_color(colors[i])

for i, patch in (enumerate(patches)):
    height = patch.get_height()
    plt.text(patch.get_x() + patch.get_width() / 2, height ,counts[i], 
             ha='center', va='bottom')
    plt.text(7,3000,"Nazira")
plt.show()
```

```python
# va='bottom': Aligns the text to the bottom of the specified y-coordinate.
# va='top': Aligns the text to the top of the specified y-coordinate.
# va='center': Centers the text vertically on the specified y-coordinate.
# va='baseline': Aligns the text to the baseline of the specified y-coordinate.

# ha='left': Aligns the text to the left of the specified x-coordinate.
# ha='right': Aligns the text to the right of the specified x-coordinate.
# ha='center': Centers the text horizontally on the specified x-coordinate.
```

```python
bins = 10
plt.xlabel('Quantity')
plt.ylabel('Count')
counts,bins,patches = plt.hist(df['Quantity'], rwidth = 0.95, bins = bins,color='gold',edgecolor = 'k',orientation = 'horizontal')
for i, patch in enumerate(patches):
    width = patch.get_width()
    plt.text(width , patch.get_y() + patch.get_height() / 2, str(int(counts[i])), 
             ha='left', va='center', fontsize=12, color='black')
```

```python
plt.figure(figsize = (10,5))

bins = 10
cmap = plt.get_cmap('terrain_r')  # Choose a colormap
colors = cmap(np.linspace(0, 1, bins))  # Generate a different color for each bin

counts,bins,patches = plt.hist(df['Quantity'], rwidth = 0.95, bins = bins,color='gold',edgecolor = 'k')

# Set colors for each bar
for i in range(len(patches)):
    patches[i].set_color(colors[i])


for i, patch in enumerate(patches):
    height = patch.get_height()
    plt.text(patch.get_x() + patch.get_width() / 2, height + 0.1, counts[i], 
             ha='center', va='bottom')
```

```python
categories = (df['Category'].value_counts(dropna = False, normalize = True) * 100).index
freq = (df['Category'].value_counts(dropna = False, normalize = True) * 100).values
freq = [round(f,2) for f in freq]
ax = plt.bar(categories, freq)
for idx, i in enumerate(ax.patches):
    x_cord = i.get_x() + i.get_width() / 2  # Center text horizontally on the bar
    y_cord = i.get_height()  # Position text at the top of the bar
    plt.text(x_cord, y_cord,f'{freq[idx]} %', ha='center', va='bottom')
plt.show()
```

```python
categories = df['Sub-Category'].value_counts().index
counts = df['Sub-Category'].value_counts().values
ax = plt.barh(categories, counts)
for i in ax.patches:
    x = i.get_width()
    y = i.get_y()
    plt.text(x, y,i.get_width(),fontdict={'fontsize' : 14})
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pyplot works in heirarchy and has access only to current axes

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

plt.figure(figsize=(6, 10))

plt.subplot(3, 1, 1)  
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc = 'upper right')

plt.subplot(312) 
plt.plot(x, y2, label='Tangent Wave', color='red')
plt.title('Tangent Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.ylim(-10, 10) 
plt.grid()
plt.legend()

plt.subplot(313) 
plt.plot(x, y3, label='Cosine Wave', color='green')
plt.title('Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend(loc = 'upper right')

# # Adjust layout and show the plot
plt.tight_layout()
# plt.show()
plt.savefig('Analysis.jpg')
```
