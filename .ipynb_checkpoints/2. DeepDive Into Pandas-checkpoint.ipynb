| **Code**  | **Description**                    |
|-----------|------------------------------------|
| `'B'`     | Business day                       |
| `'D'`     | Calendar day                       |
| `'W'`     | Weekly (Sunday by default)         |
| `'W-MON'` to `'W-SUN'` | Weekly on specific weekday |
| `'M'`     | Month end                          |
| `'SM'`    | Semi-month end (15th and end of month) |
| `'BM'`    | Business month end                 |
| `'CBM'`   | Custom business month end          |
| `'MS'`    | Month start                        |
| `'SMS'`   | Semi-month start (1st and 15th)    |
| `'BMS'`   | Business month start (first weekday of the month)          |
| `'Q'`     | Quarter end                        |
| `'BQ'`    | Business quarter end (last business day of each quarter)              |
| `'QS'`    | Quarter start                      |
| `'BQS'`   | Business quarter start (first business day of each quarter)          |
| `'A'` or `'Y'` | Year end                     |
| `'BA'` or `'BY'` | Business year end           |
| `'AS'` or `'YS'` | Year start                  |
| `'BAS'` or `'BYS'` | Business year start       |
| `'H'`     | Hourly                             |
| `'T'` or `'min'` | Minutely                    |
| `'S'`     | Secondly                           |
| `'L'` or `'ms'` | Milliseconds                 |
| `'U'` or `'us'` | Microseconds                 |
| `'N'` or `'ns'` | Nanoseconds                  |

| Frequency Code | Description                                          | Example |
|----------------|------------------------------------------------------|---------|
| `'D'`          | Day                                                 | `2023-01-01`, `2023-01-02`, etc. |
| `'B'`          | Business day (excludes weekends, Mon-Fri)            | `2023-01-01` is excluded if itâ€™s a weekend |
| `'W'`          | Week (week ending on Sunday)                       | `2023-01-01`, `2023-01-08`, etc. |
| `'M'`          | Month (end of the month)                             | `2023-01`, `2023-02`, etc. |
| `'Q'`          | Quarter (end of the quarter)                         | `2023Q1`, `2023Q2`, etc. |
| `'A'` or `'Y'` | Year (end of the year)                               | `2023`, `2024`, etc. |
| `'H'`          | Hour (hourly frequency)                              | `2023-01-01 00:00`, `2023-01-01 01:00`, etc. |
| `'T'` or `'min'` | Minute (minute-based frequency)                      | `2023-01-01 00:00:00`, `2023-01-01 00:01:00`, etc. |
| `'S'`          | Second (second-based frequency)                      | `2023-01-01 00:00:00`, `2023-01-01 00:00:01`, etc. |
| `'L'` or `'ms'` | Millisecond (millisecond-based frequency)            | `2023-01-01 00:00:00.001`, `2023-01-01 00:00:00.002`, etc. |
| `'U'` or `'us'` | Microsecond (microsecond-based frequency)            | `2023-01-01 00:00:00.000001`, `2023-01-01 00:00:00.000002`, etc. |

Extended Frequency Codes:

| Frequency Code  | Description                                              | Example |
|-----------------|----------------------------------------------------------|---------|
| `'W-MON'`       | Week starting on Monday                                  | `2023-01-02`, `2023-01-09`, etc. |
| `'W-TUE'`       | Week starting on Tuesday                                 | `2023-01-03`, `2023-01-10`, etc. |
| `'Q-JAN'`       | Quarter ending in January                                | `2023Q1`, `2023Q2`, etc. |
| `'A-JAN'`       | Year ending in January                                   | `2023`, `2024`, etc. |
| `'BM'`          | Business month (end of business month)                   | `2023-01-31`, `2023-02-28`, etc. |
| `'BQ'`          | Business quarter (end of business quarter)               | `2023Q1`, `2023Q2`, etc. |
| `'H'`           | Hourly frequency                                         | `2023-01-01 00:00`, `2023-01-01 01:00`, etc. |
| `'T'`           | Minute-based frequency                                   | `2023-01-01 00:00:00`, `2023-01-01 00:01:00`, etc. |

Tip: When reading date columns from CSV: `parse_dates` parameter auto-converts specified columns into datetime format using pd.to_datetime() behind the scenes.

#### .dt accessor

- .dt accessor in a Pandas Series allows you to access datetime-related properties and methods when your Series has datetime64[ns] or Period values.
- If the Series is of type datetime64[ns], you can extract:

    - year, month, day
    - hour, minute, second, microsecond, nanosecond
    - date â€“ returns date part only
    - time â€“ returns time part only
    - dayofweek or weekday â€“ Monday=0, Sunday=6
    - day_name() â€“ e.g., "Monday"
    - month_name() â€“ e.g., "January"
    - quarter
    - dayofyear â€“ 1 to 365/366
    - is_month_start, is_month_end
    - is_quarter_start, is_quarter_end
    - is_year_start, is_year_end
      
- If the Series contains Period values (e.g., from pd.Series(pd.period_range(...))), .dt lets you access:
    - year, month, day
    - hour, minute, etc. (if frequency includes time)
    - start_time, end_time

````python
def format_duration(seconds):
    if seconds < 60:
        return f"{seconds:.2f} sec"
    elif seconds < 3600:
        mins = int(seconds // 60)
        secs = seconds % 60
        return f"{mins} min {secs:.2f} sec"
    else:
        hours = int(seconds // 3600)
        mins = int((seconds % 3600) // 60)
        secs = seconds % 60
        return f"{hours} hr {mins} min {secs:.2f} sec"
````

````python
# If ship-date is beyond 31-Dec-2016 then delay the ship-date by a 3 month

def delay_ship_date(x):
    if x > pd.to_datetime('2016-12-31'):
        return x + pd.DateOffset(months = 3)
    else:
        return x
    

df['revised_ship_date'] = df['Ship Date'].apply(delay_ship_date)
````

#### .agg() / .aggregate()

- **Benifits of using .agg**
1. You can apply multiple aggregations at once
2. You can apply custom function to the grouped data
3. You can provide name to aggregated columns
4. You can apply different aggregations on different columns

##### 1. You can apply multiple aggregations at once

# Bollinger Bands

- Bollinger Bands are a type of technical analysis tool used to measure market volatility and identify potential overbought or oversold conditions.
- Each Bollinger Band consists of 3 lines:

    - Middle Band: This is a moving average (usually 20-period Simple Moving Average - SMA).
    - Upper Band: `UpperÂ Band = SMA + (kÃ—StandardÂ Deviation)`
    - Lower Band: `LowerÂ Band = SMA âˆ’ (kÃ—StandardÂ Deviation)`
      ```
          Where:  k is typically 2.
                  Standard deviation is based on the last N periods.
      ```

Interpretation:
- When the price touches the upper band, it may be overbought.
- When the price touches the lower band, it may be oversold.
- If the bands are wide apart, high volatility. If close together, low volatility.



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data (You can replace this with your own stock data)
dates = pd.date_range(start="2024-01-01", periods=100)
np.random.seed(0)
price = np.cumsum(np.random.randn(100)) + 100  # simulated price

df = pd.DataFrame({'Price': price}, index=dates)

# Calculate Bollinger Bands
window = 20
df['SMA'] = df['Price'].rolling(window=window).mean()
df['STD'] = df['Price'].rolling(window=window).std()
df['Upper'] = df['SMA'] + (2 * df['STD'])
df['Lower'] = df['SMA'] - (2 * df['STD'])

# Plotting
plt.figure(figsize=(12, 4))
plt.plot(df['Price'], label='Price')
plt.plot(df['SMA'], label='SMA (20)')
plt.plot(df['Upper'], label='Upper Band', linestyle='--')
plt.plot(df['Lower'], label='Lower Band', linestyle='--')
plt.fill_between(df.index, df['Lower'], df['Upper'], color='gray', alpha=0.1)
plt.title('Bollinger Bands')
plt.legend()
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(filepath_or_buffer='./data/GOOGL.csv', parse_dates=True)
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'].dt.year > 2022].set_index('Date')
df
````

````python
# Calculate Bollinger Bands
window = 21
df['SMA'] = df['Close'].rolling(window=window).mean()
df['STD'] = df['Close'].rolling(window=window).std()
df['Upper'] = df['SMA'] + (2 * df['STD'])
df['Lower'] = df['SMA'] - (2 * df['STD'])

# Plotting
fig = plt.figure(figsize=(14, 6))
fig.suptitle('Google -- 2 stds above/below closing price')
plt.plot(df['Close'], label='Price')
plt.plot(df['SMA'], label='SMA (20)')
plt.plot(df['Upper'], label='Upper Band', linestyle='--')
plt.plot(df['Lower'], label='Lower Band', linestyle='--')
plt.fill_between(df.index, df['Lower'], df['Upper'], color='gray', alpha=0.1)
plt.title('Bollinger Bands')
plt.legend()
plt.show()
````
